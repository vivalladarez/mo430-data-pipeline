```python
from __future__ import annotations

import sys
from pathlib import Path


def repo_root() -> Path:
    here = Path.cwd().resolve()
    for p in [here, *here.parents]:
        if (p / "dags" / "utils" / "data_cleaners.py").is_file():
            return p
    raise RuntimeError("Defina o cwd na raiz do repo ou em dags/medallion; repo_root procura dags/utils/data_cleaners.py.")


ROOT = repo_root()
sys.path.insert(0, str(ROOT / "dags"))

from medallion.silver.silver import (
    SILVER_GEO_NODES_CSV,
    SILVER_GEO_NODES_PRINCIPAL_CSV,
    run_silver_geo_nodes,
    run_silver_geo_nodes_principal,
)

silver_dir = ROOT / "data" / "silver"
ROOT
```
